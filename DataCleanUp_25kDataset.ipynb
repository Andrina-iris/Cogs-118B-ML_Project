```python
import pandas as pd
import numpy as np

#Upload File
df = pd.read_csv("25k.csv", sep=",")

df
```

```python
#Add zero infront of single digit days & months
df['Day'] = (df['Day'].astype(str)).apply(lambda x: x.zfill(2))
df['Month'] = (df['Month'].astype(str)).apply(lambda x: x.zfill(2))

#Combine Year, Month, Day into Date column
#Date Format: YYYY-MM-DD
df['Date'] = df['Year'].astype(str) + '-' + df['Month'].astype(str) + '-' + df['Day'].astype(str)
df.insert(0,'Date', df.pop('Date'))

#Drop columns
df = df.drop(['Year', 'Month','Day',"Timestamp"], axis = 1)
df
```

```python
df.to_csv('clean25kDataset', index = False)
```
